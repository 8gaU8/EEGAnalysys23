```python
import matplotlib

# matplotlib.use("qt5agg")
import mne
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
root = Path("/Volumes/data/haga/data/inter_files/avg_erps/")
avg_move_m = mne.read_evokeds(root / "avg_move_m-ave.fif.gz", verbose=False)[0]
avg_move_nm = mne.read_evokeds(root / "avg_move_nm-ave.fif.gz", verbose=False)[0]
avg_no_move_m = mne.read_evokeds(root / "avg_no_move_m-ave.fif.gz", verbose=False)[0]
avg_no_move_nm = mne.read_evokeds(root / "avg_no_move_nm-ave.fif.gz", verbose=False)[0]
```

```python
picks = ["Fpz", "Fp1", "Fp2", "AF7", "AF3", "AF4", "AF8", "Fz", "T7", "Cz", "T8", "Pz", "Oz"]
```

# Don't Move, ミュージシャン

```python
avg_no_move_m.plot(
    gfp=True,
    spatial_colors=True,
    picks=picks,
    window_title="avg_no_move_m",
)
```

# Don't Move, 非ミュージシャン

```python
avg_no_move_nm.plot(
    gfp=True,
    spatial_colors=True,
    picks=picks,
    window_title="avg_no_move_nm",
)
```

# Move, ミュージシャン

```python
avg_move_m.plot(
    gfp=True,
    spatial_colors=True,
    picks=picks,
    window_title="avg_move_m",
)
```

# Move, 非ミュージシャン

```python
avg_move_nm.plot(
    gfp=True,
    spatial_colors=True,
    picks=picks,
    window_title="avg_no_move_nm",
)
```
